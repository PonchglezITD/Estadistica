```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Cargar el dataset
file_path = "/content/autos2.csv"
autos = pd.read_csv(file_path)
```

```python
# Mostrar las primeras filas del dataset
autos.head()
```

```python
# Mostrar la informacion del archivo.
# Muestra el numero de registros, tipos de datos de la columna, valores nulos, memoria utilizada
autos.info()
```

```python
# Medidas de tendencia central
min_price = autos['price'].min()
max_price = autos['price'].max()
mode_price = autos['price'].mode()[0]
mean_price = autos['price'].mean()
median_price = autos['price'].median()
std_price = autos['price'].std()
describe_price = autos['price'].describe()
```

```python
# Imprimir valores
print(f"Precio mínimo: {min_price}")
print(f"Precio máximo: {max_price}")
print(f"Moda del precio: {mode_price}")
print(f"Media del precio: {mean_price}")
print(f"Mediana del precio: {median_price}")
print(f"Desviación estándar del precio: {std_price}")
print("\nEstadísticos y cuartiles con la funcion describe():")
print(describe_price)
```

#automóviles


```python
# Histograma de precios
plt.figure(figsize=(10,6))
sns.histplot(autos['price'], bins='sturges', kde=True, color='purple')
plt.title("Distribución de Precios de Automóviles")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.show()

print("Conclusión: ")
print("La mayoría de los autos tienen precios bajos, con algunos valores altos poco frecuentes.\nLa distribución está sesgada a la derecha, indicando que hay menos autos de lujo o con precios elevados.")
```

```python
# Diagrama de caja y bigotes - Precio y Tipo de Vehículo
plt.figure(figsize=(10,6))
sns.boxplot(x='body-style', y='price', color='purple',data=autos)
plt.title("Relación entre Precio y Tipo de Vehículo")
plt.xlabel("Tipo de Vehículo")
plt.ylabel("Precio")
plt.xticks(rotation=0)
plt.show()

print("Conclusión: ")
print("Este diagrama de caja muestra cómo varían los precios entre los diferentes tipos de vehículos.\nPodemos observar que ciertos tipos tienen una mayor dispersión de precios y posibles valores atípicos.")

```

```python
# Diagrama de caja y bigotes - Precio y Tipo de Tracción
plt.figure(figsize=(10,6))
sns.boxplot(x='drive-wheels', y='price', color='purple', data=autos)
plt.title("Relación entre Precio y Tipo de Tracción")
plt.xlabel("Tipo de Tracción")
plt.ylabel("Precio")
plt.show()

print("Conclusión:")
print("Aquí se puede analizar cómo la tracción afecta el precio del vehículo.\nSe puede notar que ciertos tipos de tracción tienden a tener precios más altos en promedio.")
```

```python
# Diagrama de caja y bigotes - Precio y Ubicación del Motor
plt.figure(figsize=(10,6))
sns.boxplot(x='engine-location', y='price', color='purple', data=autos)
plt.title("Relación entre Precio y Ubicación del Motor")
plt.xlabel("Ubicación del Motor")
plt.ylabel("Precio")
plt.show()

```
